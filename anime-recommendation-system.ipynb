# Different Methods of Recommendation
1. Best Rated
2. Highest Number of Members
3. Recommendation based on Genre, Show, Type, etc

# Boring Methods

## Best Rated
This will always show the highest rated shows 

```python
num_of_shows = 15
top_shows = anime_list.sort_values(by=["rating"], ascending=False)['name'].head(num_of_shows)
for each in top_shows:
    print(each)
```

## Highest Number of Members
This will always show the anime with the most members to show what has been watched the most

```python
most_watched_shows = anime_list.sort_values(by=["members"], ascending=False)['name'].head(num_of_shows)
for each in most_watched_shows:
    print(each)
```
